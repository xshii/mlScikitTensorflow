## Voting classifiers - `ensemble:VotingClassifier`

### hard voting classifier  - `voting='hard'` (by default)

The majority-vote classifier is called a *hard voting* classifier.

In fact, even if each classifier is a *weak learner* ( meaning it does only slightly better than random guessing), the ensemble can still be a *strong learner* (achieve high accuracy), **provided there are a sufficient number of weak learners and they are sufficiently diverse.**

example:

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
data = make_moons(n_samples=300,random_state=42,noise=0.5)
X_train, X_test = train_test_split(data[0],random_state=42)
y_train, y_test = train_test_split(data[1],random_state=42)
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
```

```python
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC()
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard'
)
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

## Bagging and Pasting

Use the same training algorithm for every predictor, but to train them on different random subsets of the training set. It has a comparable bias but a smaller variance.

**bagging**: when sampling is performed with replacement (short for *bootstrap aggregating*)

**pasting**: when sample................without.........

Predictors can all be trained in parallel, via different CPU cores or even different servers.

Similarly, predictions can be made in parallel.

### Bagging & pasting in SL - `ensemble:BaggingClassifier(bootstrap=True/False)`

```python
from sklearn.ensemble import BaggingClassifier 
from sklearn.tree import DecisionTreeClassifier
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=200, bootstrap=True, n_jobs=-1
)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

bagging ends up with a slightly higher bias than pasting, but this also means that predictors end up being less correlated so the ensemble’s variance is reduced. Overall, bagging often results in better models, which explains why it is generally preferred.

### Out-of-Bag Evaluation - `oob_score=True`, access by `.oob_score_`

When `bootstrap=True`, 37% of the training instances are not sampled which are called *out-of-bag* instances. Note that they are not the same 37% for all predictors.

You can evaluate the ensemble itself by averaging out the oob evaluations of each predictor.

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500, bootstrap=True, n_jobs=-1,oob_score=True)
bag_clf.fit(X_train, y_train)
```

```python
bag_clf.oob_score_
```

```python
from sklearn.metrics import accuracy_score
y_pred = bag_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
bag_clf.oob_decision_function_
```

### Random patches and random subspaces - `max_features` and `bootstrap_features`

The `BaggingClassifier` class supports sampling the features as well.

Sampling features results in even more predictor diversity, trading a bit more bias for a lower variance

####  Random patches method - `bootstrap=True, bootstrap_features=True` or `max_features=0.63`

**Random patches method**: sampling both training instances and features

#### Random subspaces method - `bootstrap=False, max_samples=1.0, bootstrap_features=True` or `max_features=0.63`

**Random subspaces method**: keeping all training instances but sampling features

## Random Forests - `ensemble:RandomForestClassifier/RandomForestRegressor`

The following code trains a Random Forest classifier with 500 trees (each limited to maximum 16 nodes), using all available CPU cores.

```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1)
rnd_clf.fit(X_train, y_train)
y_pred_rf = rnd_clf.predict(X_test)
print(sum(y_pred_rf==y_test)/len(y_test))
```

The `RandomForestClassifier` has 
* all the hyperparameters of a `DecisionTreeClassifier` plus 
* all the hyperaparameters of a `BaggingClassifier` to control the ensemble itself.

Instead of searching for the very best feature when splitting a node, it searches for the best feature among a random subset of features. This results in a greater tree diversity, which (once again) trades a higher bias for a lower variance, generally yielding an overall better model. 

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(splitter="random", max_leaf_nodes=16),
    n_estimators=500, max_samples=1.0, bootstrap=True, n_jobs=-1
)
```

### Extra-trees - `ensemble:ExtraTreesClassifier/ExtraTreesRegressor`

**Extremely randomized trees** ensemble: using random thresholds for each features rather than searching for the best possible thresholds. 

This trades more bias for a lower variance.

It also makes extra-trees much faster to train than regular random forests since finding the best possible threshold for each feature at every node is one of the most time-consuming tasks of growing a tree.

### Feature importance `.feature_importance_`

Important features are likely to appear closer to the root of the tree, while unimportant feature will often appear closer to the leaves (or not at all).

`.feature_importance_`: the average depth. It is computed by SL automatically for every feature after training.


```python
from sklearn.datasets import load_iris
iris = load_iris()
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=-1)
rnd_clf.fit(iris["data"], iris["target"])
for name, score in zip(iris["feature_names"], rnd_clf.feature_importances_):
    print(name, score)
```

## Boosting

*Boosting* (originally called *hypothesis boosting*) refers to any Ensemble method that can combine several weak learner into a strong learner.

The general idea of boosting methods is to train predictors sequentially, each trying to correct its predecessor.

The most popular boosting method are
* *AdaBoost* (Adaptive Boosting)
* *Gradient Boosting*

### AdaBoost

**AdaBoost**: The new predictor pays a bit more attention to the **underfitted** training instances. This results in new predictors focusing more on the hard cases.

For example, to build an AdaBoost classifier, after the base classifier, the relative weight of misclassified training instances is then increased....

The ensemble makes predictions very much like bagging or pasting, except that predictors have different weights depending on their overall accuracy on the weighted training set.

1. Each instance weight $w^{(i)}$ is initially set to $\frac{1}{m}$. A first predictor is trained and its weighted error rate $r_1$ is computed on the training set by 

$$r_j=\frac{\sum_{i=1,\hat{y}_j^{(i)}\neq y^{(i)}}^mw^{(i)}}{\sum_{i=1}^{m}w^{(i)}}$$

where $\hat{y}_j^{(i)}$ is the $j^{th}$ predictor's prediction for the $i^th$ instance

2. The predictor's weight $\alpha_j$ is then computed using
$$\alpha_j=\eta\log\frac{1-r_j}{r_j}$$

where $\eta$ is the learning rate hyperparameter (default to 1). The more accurate the predictor is, the higher its weight will be. Random guess leads to 0. Worse predictor leads to negative number.

3. The instance weights are updated using (for $i=1,2,...,m$)
    $$w^{(i)}=w^{(i)}\, \text{ if } \hat{y}_j^{(i)}=y^{(i)}$$
$$w^{(i)}=w^{(i)}\exp(\alpha_j)\, \text{ if } \hat{y}_j^{(i)}\neq y^{(i)}$$

4. Then all the instance weight are normalized (divided by $\sum_{i=1}^{m}w^{(i)}$_

5. A new predictor is trained using the updated weights. Stop when 
    * the desired number of predictors is reached or
    * a perfect predictor is found
    
    or Return to 1.

To make predictions, AdaBoost simply computes the predictions of all the predictors and wieghts them using the predictor weights $\alpha_j$. The predicted class is the one that receives the majority of weighted votes

$$\hat{y}(x)=\arg\max_k\sum^N_{j=1,\hat{y}_j(x)=k}\alpha_j$$

where $N$ is the number of the predictors

#### SL SAMME - `ensemble:AdaBoostClassifier/AdaBoostRegressor`

**SAMME**(Stagewise Additive Modeling using a Multiclass Exponential loss function): a multiclass version of AdaBoost.

**SAMME.R** (Real), when the predictors can estimate class probabilities (have a `predict_proba()` method.

**Decision Stumps**: a Decision Tree with `max_depth=1`, this is the default base predictor for the `AdaBoostClassifier`

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200,
    algorithm='SAMME.R', learning_rate=0.5
)

ada_clf.fit(X_train,y_train)
y_pred = ada_clf.predict(X_test)
print(sum(y_pred==y_test)/len(y_pred))
```

### Gradient Boosting

Gradient boosting method tries to fit the new predictor to the *residual errors* made by the previous predictor

**GBRT**(Gradient Boosted Regression Trees): using Decision Trees as the base predictors.

An example of GBRT:

```python
import numpy as np
X = np.linspace(-3,2,100)
y = X**2 + 2*X -1 + np.random.rand(len(X))

plt.scatter(X,y)
```

```python
from sklearn.tree import DecisionTreeRegressor 
tree_reg1 = DecisionTreeRegressor(max_depth=2)
X = X.reshape(-1,1)
tree_reg1.fit(X, y)
```

Fit the residual error to next predictors

```python
y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2)
tree_reg2.fit(X, y2)
y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2)
tree_reg3.fit(X, y3)
```

```python
X_new = np.linspace(-2,2,50).reshape(-1,1)
y_pred = sum(tree.predict(X_new) for tree in (tree_reg1, tree_reg2, tree_reg3))
plt.scatter(X_new,y_pred)
```

####  SL GBRT - `ensemble:GradientBoostingRegressor`

```python
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0)
gbrt.fit(X,y)
```

#### Find the optimal number of trees - `.staged_predict()`

```python
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error
    
X_train, X_val, y_train, y_val = train_test_split(X, y)
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=120)
gbrt.fit(X_train, y_train)
errors = [mean_squared_error(y_val, y_pred) for y_pred in gbrt.staged_predict(X_val)]

bst_n_estimators = np.argmin(errors)
gbrt_best = GradientBoostingRegressor(max_depth=2,n_estimators=bst_n_estimators)
gbrt_best.fit(X_train, y_train)
```

```python
plt.scatter(np.arange(len(errors)),errors)
```

```python
gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True)

min_val_error = float("inf")
error_going_up = 0
for n_estimators in range(1, 120):
    gbrt.n_estimators = n_estimators 
    gbrt.fit(X_train, y_train)
    y_pred = gbrt.predict(X_val)
    val_error = mean_squared_error(y_val, y_pred)
    if val_error < min_val_error:
            min_val_error = val_error
            error_going_up = 0
    else:
        error_going_up += 1
        if error_going_up == 5:
            break # early stopping
```

`subsample=0.25` use a subset of training set to speed up training considerably. This technique is call *Stochastic Gradient Boosting*

## Stacking

**Stacking** (stacked generalization): instead of using trivial functions (such as hard voting) to aggregate the predictions of all predictors in an ensemble, a model (called a *blender* or a `meta learner`) is trained to perform this aggregation.

SL does not support stacking directly. [brew](https://github.com/viisar/brew)
